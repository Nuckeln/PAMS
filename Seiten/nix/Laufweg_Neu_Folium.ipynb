```python
import pandas as pd

df = pd.read_excel('Lager.xlsx')

# Erstelle eine neue Spalte mit der fortlaufenden Nummer
df['H'] = range(1, len(df) + 1)

# Erstelle eine neue Spalte, die angibt, ob sich die 3. Stelle in Column A verändert hat
df['A_changed'] = df['Stellplatz'].apply(lambda x: str(x)[2])

# Erstelle eine neue Spalte, die angibt, ob sich die 3. Stelle in Column A verändert hat
df['H_changed'] = df['H'].diff().fillna(0) != 1

# Erstelle eine neue Spalte, die angibt, ob eine neue Spalte eingefügt werden muss
df['Insert_column'] = df['A_changed'] != df['A_changed'].shift(1) | df['H_changed']
df['Insert_column'].fillna(False, inplace=True)  # Fülle NaN-Werte mit False

# Füge eine neue Spalte ein, wenn erforderlich
for i, row in df.iterrows():
    if row['Insert_column']:
        df.insert(i, 'Neue Spalte', '')

# Speichere die Daten in einer neuen Excel-Datei
df.to_excel('Lager_neu.xlsx', index=False)
```
